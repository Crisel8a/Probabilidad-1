# VARIABLES ALEATORIAS, DISTRIBUCIONES, ESPERANZA Y VARIANZA

## Distribución Binomial 
**Función de densidad**
$$P[X = x] = {n \choose k} p^k (1-p){n-k}$$


**Esperanza y Varianza**
$$E(X) = np, Var(X) = p(1-p)$$

## Distribución Binomial Negativa
**Función de masa**
$$P[X = k] = {k+r-1 \choose k} (1-p)^k p^r$$

**Esperanza y Varianza**
$$E(X) = \frac{rq}{p}, Var(X) = \frac{rq}{p^2}$$

## Distribución Bernoulli
**Función de densidad**
$$P[X = x] = p^x(1-p)^{1-x}, x \in [0,1]$$
**Función de Distribución**
$$F(X) = \begin{cases} 0, & \mbox{si } x<0 \\ 1-p, & \mbox{si } o \leq x < 1 \\ 1, & \mbox{si } x \geq 1 \end{cases} $$
**Esperanza y Varianza**
$$E(X) = p, Var(X) = pq$$

## Distribución Poisson
**Función de densidad**
$$P[X = k] = e^{- \lambda} \frac{\lambda^k}{k!}$$

**Esperanza y Varianza**
$$E(X) = p, Var(X) = pq$$

## Distribución Geométrica
**Función de masa**
$$P[X = k] = p(1-p)^k$$

**Esperanza y Varianza**
$$E(X) = \frac{q}{p}, Var(X) = \frac{q}{p^2}$$

## Distribución Exponencial
**Función de masa**
$$P[X = x] = \lambda e^{- \lambda x}$$
**Función de Distribución**
$$F(X) = 1- e^{- \lambda x}$$
**Esperanza y Varianza**
$$E(X) = \frac{1}{\lambda}, Var(X) = \frac{1}{\lambda^2}$$

```python
from random import randint
N = 10000
cuentas = [0 for i in range(N)]
E = 0
for i in range(N):
    suma = randint(1,6) + randint(1,6)
    cuentas[suma] += 1

for xi, ni in zip(cuentas, range(N)):
    E += (ni*xi)/N

E
```
