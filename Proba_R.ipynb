## Simular lanzamiento de dos monedas

```r
#Simular lanzamiento de dos monedas
#consideramos 0=sol y 1=águila
sample(0:1, 2,replace=TRUE,c(0.75,0.25)) #rbinom(2,2,0.5)
#Para guardar las repeticiones de este experimento usamos la función replicate
numRepeticiones = 10
repeticiones = replicate(n = numRepeticiones,{
  partida = sample(0:1, size = 2, replace = TRUE)
})
class(repeticiones)
dim(repeticiones)
repeticiones=t(repeticiones) ##la transponemos

#si queremos estudiar el evento en que sale al menos un águila
#repeticiones == 1 #Esto se puede omitir ya que sólo hay dos posibles resultados en cada entrada pero es necesario si se tiene un experimento con más opciones
(filasConAguila = (rowSums(repeticiones) > 0)) #con binom--> AlgunaAguila=(rbinom(10,2,0.5)>0)

#Calculamos la frecuencia de aparición 
sum(filasConAguila)/numRepeticiones #con binom--> sum(AlgunaAguila)/numRepeticiones
```

```r
#Ejercicio: Calcular la proporción de veces en que salen dos águilas
(filasDosAguila = (rowSums(repeticiones) ==2))
sum(filasDosAguila)/numRepeticiones
```

```r
#Vamos a ver cómo la frecuencia de aparición de dos águilas se aproxima a un valor determinado
numRepeticiones = 100 #00 primero con 10 para ver qué pasa
repeticiones = replicate(n = numRepeticiones,{
  partida = sample(0:1, size = 2, replace = TRUE)
})
dim(repeticiones)
repeticiones=t(repeticiones) ##la transponemos
(filasDosAguila = (rowSums(repeticiones) ==2)) 
fAcumulada2aguilas=(cumsum(filasDosAguila))/1:100 
head(fAcumulada2aguilas)                        
tail(fAcumulada2aguilas)

plot(fAcumulada2aguilas)
lines(fAcumulada2aguilas)
abline(h=0.25)
```

```r
#Ejercicio: Obterner la gráfica del número de veces que se obtiene un águila
(filasUnAguila = (rowSums(repeticiones) ==1))
fAcumuladaUNaguila=(cumsum(filasUnAguila))/1:100
plot(fAcumuladaUNaguila)
```

# CUMPLEANOS


**¿cuál es la probabilidad de que en un cuarto con n personas, dos compartan cumpleanos**

P(ninguna coindica) = $C^c$ $\leq$ $C$ = P(al enos una coincida)


$\Omega$ = { $(x_1,...,x_n)$ : $x_i \in$ {1,...,365}}

$$P(C^c) =  \frac{365!}{365^n (365-n)!} $$

```r
p <- sample(1:365, 80, replace=TRUE)

length(p[duplicated(p)])
```

```r
sum(duplicated(p))
```

```r
plot(p)
```

```r
cumple <- function(n)
    exp(lfactorial(365)-lfactorial(365-n)-n*log(365))
```

```r
cumple(80)
```
